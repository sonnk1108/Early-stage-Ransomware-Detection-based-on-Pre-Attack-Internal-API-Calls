```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('esrd_multiclass_dataset.csv')
df['row_sum'] = df.apply(lambda row: row.sum(), axis=1)
```

```python
df
```

```python
api_category_dict = {
    'Process': ['CreateProcessInternalW', 'Process32NextW', 'NtCreateFile', 'NtTerminateProcess', 'NtResumeThread', 'WriteProcessMemory', 'LdrGetDllHandle'],
    'System Info': ['GetComputerNameA', 'GetComputerNameW', 'GlobalMemoryStatusEx', 'NtQuerySystemInformation', 'IsDebuggerPresent'],
    'Memory': ['NtAllocateVirtualMemory', 'NtMapViewOfSection', 'NtProtectVirtualMemory', 'NtUnmapViewOfSection'],
    'Registry': ['RegOpenKeyExA', 'RegOpenKeyExW', 'RegSetValueExA', 'RegSetValueExW', 'NtCreateKey', 'NtSetValueKey'],
    'File System': ['MoveFileWithProgressW', 'SetFileAttributesW', 'NtWriteFile', 'GetDiskFreeSpaceExW', 'GetDiskFreeSpaceW'],
    'Services': ['CreateServiceA', 'CreateServiceW', 'EnumServicesStatusA', 'EnumServicesStatusW'],
    'Network': ['GetAdaptersAddresses', 'InternetOpenA'],
    'GUI Interactions': ['FindWindowA', 'ShellExecuteExW', 'WriteConsoleA', 'WriteConsoleW'],
    'Privileges': ['LookupPrivilegeValueW'],
    'Devices': ['DeviceIoControl'],
    'Cryptography': ['CryptExportKey', 'CryptGenKey'],
    'Threads': ['NtGetContextThread', 'NtSetContextThread']
}

# Add new columns and calculate the sum for each row
for category, api_list in api_category_dict.items():
    df[category] = df[api_list].sum(axis=1)

# Calculate percentage for each new column relative to the total value of the row
for category in api_category_dict.keys():
    df[f'{category} (%)'] = (df[category] / df.sum(axis=1)) * 100
```

```python
df.to_csv('output_processed.csv', index=False)
```
